## What is SNR
Signal-to-noise ration is defined as the ratio of the power of a signal to the power of noise.

### SNR=P(signal)/P(noise)

### SNR dB=10*log10[P(signal)/P(noise)]

### SNR dB=10*log10(Psignal) - 10*log10(Pnoise)

```python
# import libraries files numpy and matplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Genrate 1000 samples from 1 to 100
t = np.linspace(1, 100, 1000)
signal = 10*np.sin(t/(2*np.pi))

plt.plot(t, signal)
plt.title('Signal')
plt.ylabel('Magnitude')
plt.xlabel('Time (s)')
plt.show()
```

```python
# power the signal
power = signal ** 2
plt.plot(t, power)
plt.title('Signal Power')
plt.ylabel('Power (W)')
plt.xlabel('Time (s)')
plt.show()
```

Find the signal power in term of `Decibel`

```python
# convert power in dB
signal_power_db = 10 * np.log10(power)
plt.plot(t, signal_power_db)
plt.title('Signal Power in dB')
plt.ylabel('Power (dB)')
plt.xlabel('Time (s)')
plt.show()
```

```python
# add SNR of 20 dB
snr_db = 20

# Calculate signal power
signal_average_power = np.mean(power)

# convert signal power to dB
signal_average_power_db = 10 * np.log10(signal_average_power)

# Calculate noise using -> SNR dB=10*log10(Psignal) - 10*log10(Pnoise)
noise_db = signal_average_power_db - snr_db

# Convert noise from db to watts
noise_watts = 10 ** (noise_db/10)

# Genrate an sample of white noise
mean_noise = 0
noise = np.random.normal(mean_noise, np.sqrt(noise_watts), len(signal))

# Noise added to the original signal
noise_signal = signal + noise
```

```python
# Plot noise signal
plt.plot(t, noise_signal)
plt.title('Signal with noise')
plt.ylabel('Magnitude')
plt.xlabel('Time (s)')
plt.show()
```
